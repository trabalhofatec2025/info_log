# Task
Analyze the "ice creem.csv" file by performing the following steps: load the data into a pandas DataFrame, generate a scatter plot of revenue versus temperature, and create a linear regression model to predict revenue based on temperature.

## Prepare the data

### Subtask:
Separate the data into features (Temperature) and the target variable (Revenue).


**Reasoning**:
Separate the data into features (Temperature) and the target variable (Revenue) as instructed.



```python
X = df['Temperature'].values.reshape(-1, 1)
y = df['Revenue']
```

## Split the data

### Subtask:
Split the data into training and testing sets.


**Reasoning**:
Split the data into training and testing sets using train_test_split.



```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Train the model

### Subtask:
Train a linear regression model using the training data.


**Reasoning**:
Import the LinearRegression class, instantiate the model, and fit it to the training data.



```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

## Evaluate the model

### Subtask:
Evaluate the model's performance using the testing data.


**Reasoning**:
Evaluate the model's performance using the testing data by calculating MSE and R-squared.



```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

## Visualize the results

### Subtask:
Plot the regression line on the scatter plot.


**Reasoning**:
Generate a scatter plot of the testing data and plot the regression line on the same plot with labels, a title, and a legend.



```python
plt.scatter(X_test, y_test, label='Dados de Teste')
plt.plot(X_test, y_pred, color='red', label='Linha de Regressão')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Vendas ($)')
plt.title('Grafico de dispersão com linha de regressão')
plt.legend()
plt.show()
```
